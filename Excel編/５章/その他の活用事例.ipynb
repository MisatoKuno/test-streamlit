```python
import pandas as pd

df=pd.read_csv("問い合わせ履歴_4月末時点.csv",encoding="shift-jis")
df2=pd.read_excel("所属部署リスト.xlsx",sheet_name="Sheet1")

df_merge=pd.merge(df,df2,on="社員番号")
df_merge=df_merge.iloc[:,[0,1,2,9,3,4,5,6,7]].sort_values("ID")

df_merge.to_csv("問い合わせ履歴_4月末時点_部署追記.csv",encoding="shift-jis",index=False)
```

```python
df_merge2=pd.merge(df,df2,left_on="担当者",right_on="名前")
df_merge2=df_merge2.rename(columns={"社員番号_x":"社員番号"})
df_merge2
```

# 差分検出

```python
import pandas as pd
df_4=pd.read_csv("問い合わせ履歴_4月末時点.csv",encoding="shift-jis")
df_5=pd.read_csv("問い合わせ履歴_5月末時点.csv",encoding="shift-jis")

df_merge=pd.merge(df_4,df_5,on="ID",how="outer",indicator=True)
df_merge
```

# ピポットテーブル

```python
import pandas as pd

df=pd.read_csv("問い合わせ履歴_4月末時点.csv",encoding="shift-jis")

df.pivot_table(index="担当者",columns="取引先",values="対応工数(h)",aggfunc="sum")
```

```python
df_p=df.pivot_table(index="担当者",columns=["取引先","カテゴリ"],values="対応工数(h)",aggfunc="sum",fill_value=0)
df_p.to_excel("ピボットテーブル.xlsx")
```
