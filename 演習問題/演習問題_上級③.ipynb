## *テーマ：コラッツ予想*

コラッツ予想とは、下記のような数学の未解決問題です。（出典：Wikipedia）
https://ja.wikipedia.org/wiki/%E3%82%B3%E3%83%A9%E3%83%83%E3%83%84%E3%81%AE%E5%95%8F%E9%A1%8C
 <br>
> 「任意の正の整数nに対して以下の操作を繰り返した時、必ず1に到達する」という予想
> - nが偶数の場合、nを2で割る
> - nが奇数の場合、nに3をかけて1を足す

  
例えば、5に対して上記の操作を繰り返すと、「5→16→8→4→2→1」となります。
<br>
<br>

**問１．  
2以上100以下の整数に対して、コラッツ予想が成り立つかどうか、および各整数について1に到達するまでに何回の操作を要するかを調べてください。**

```python
# 回答欄
List=list(range(2,101))

for i in List:
    num=i
    count=0
    for j in range(10000):
        if num==1:
            break
        elif num%2==0:
            num=num/2
            count+=1
        else:
            num=num*3+1
            count+=1
    print("{}は{}回の操作で1になります。".format(i,count))
        
print("計算結果は全て「1」になりました。\nコラッツ予想は成立しました。")
```
