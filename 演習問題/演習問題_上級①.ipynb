**問１．  
10以上100以下の素数を順番にprintしてください。  ※素数：2以上の自然数で、正の約数が1と自分自身のみであるもの**

```python
# 回答欄
List=list(range(10,101))

for i in List:
    yakusu=[]
    for j in range(1,i+1):
        if i%j==0:
            yakusu.append(j)
    if len(yakusu)<3:
        print(i)
```

```python
List=list(range(10,101))

for i in List:
    flag=0
    for j in range(2,i):
        if i%j==0:
            flag=1
            break
    if flag==0:
        print(i)
```

**問２．  
10以上100以下の整数について、順番に以下のようにprintしてください。**
- その整数が素数の場合は、その整数をそのままprint  
- その整数が素数でない場合は、その整数の全ての正の約数を格納したリストをprint

```python
# 回答欄
List=list(range(10,101))

for i in List:
    yakusu=[]
    for j in range(1,i+1):
        if i%j==0:
            yakusu.append(j)
    if len(yakusu)<3:
        print(i)
    else:
        print(yakusu)
```
