```python
import pandas as pd
df=pd.read_excel("売上一覧.xlsx",sheet_name=0)

List=["田中","佐藤","鈴木","中村"]

for i in List:
    df_tantou=df[df["担当者"]==i]
    df_tantou.to_excel("売上一覧_{}.xlsx".format(i),index=False)
```

# ファイルの結合

```python
from glob import glob
files=glob("売上一覧_*.xlsx")

df_nakamura=pd.read_excel(files[0],sheet_name="Sheet1")
df_satou=pd.read_excel(files[1],sheet_name="Sheet1")

df_concat=pd.concat([df_nakamura,df_satou])
df_concat
```

```python
from glob import glob
import pandas as pd
files=glob("売上一覧_*.xlsx")

df_concat=pd.DataFrame()

for file in files:
    df_tantou=pd.read_excel(file,sheet_name="Sheet1")
    df_concat=pd.concat([df_concat,df_tantou])
df_concat.sort_values("No.").to_excel("売上一覧_結合.xlsx",index=False)
```
